```python
import google.generativeai as genai
from telegram import Update
from telegram.ext import ApplicationBuilder, CommandHandler, MessageHandler, filters, ContextTypes
import nest_asyncio
import asyncio

nest_asyncio.apply()

genai.configure(api_key=GEMINI_API_KEY)
model = genai.GenerativeModel("gemini-2.0-flash-001")
```

```python
# Память чата для каждого пользователя
chat_history = {}
MAX_HISTORY = 10 # Define MAX_HISTORY here

async def start(update: Update, context: ContextTypes.DEFAULT_TYPE):
    await update.message.reply_text("Привет! 🤖 Я бот на Google Gemini. Задай мне любой вопрос!")

async def handle_message(update: Update, context: ContextTypes.DEFAULT_TYPE):
    user_id = update.message.chat_id
    text = update.message.text

    if user_id not in chat_history:
        chat_history[user_id] = []

    # Добавляем сообщение пользователя
    chat_history[user_id].append({"role": "user", "parts": [text]})

    # Limit chat history length
    if len(chat_history[user_id]) > MAX_HISTORY:
        chat_history[user_id] = chat_history[user_id][-MAX_HISTORY:]

    await update.message.reply_text("⏳ Думаю...")

    # Генерируем ответ Gemini
    response = model.generate_content(chat_history[user_id])
    reply_text = response.candidates[0].content.parts[0].text

    # Добавляем ответ модели в историю
    chat_history[user_id].append({"role": "model", "parts": [reply_text]})

    # Отправляем ответ, разбивая его на части если необходимо
    max_message_length = 4096  # Maximum message length in Telegram
    for i in range(0, len(reply_text), max_message_length):
        await update.message.reply_text(reply_text[i:i+max_message_length])
```
